```
import pandas as pd
url='https://raw.githubusercontent.com/mansisinha/taddhita/master/allPairs.csv'

df=pd.read_csv(url)
df
df.groupby('affix').count()
```

```
import pandas as pd
df2=pd.read_csv('taddhita/newPairs.txt',header=None)
df2
```

```
df2.groupby(2).count()
```

```
df2['n']=0
df2
```

```
df['indf']=1
df3=pd.merge(df2,df,how='left',left_on=[0,1,2],right_on=['Source Word','Derived Word','Affix'])
df3['n']=df3['n']+df3['indf'].fillna(0)
df3.drop(['Source Word','Derived Word','Affix','affix','word','sense','meaning','Source Word Count','Derived word count','word_y','indf'],axis=1,inplace=True)
df3.columns=['source_word','derived_word','affix','n']
df3
```

```
df3.groupby('n').count()
```

```
df3['counts']=df3.groupby('affix')['n'].transform('count')
df3['m']=1
cols=['affix','n','counts']
cols1=['m']
df3[['{}_sum'.format(x) for x in cols1]] = df3.groupby(cols)[cols1].transform('sum')
df3=df3.sort_values('counts')
df3
```

```

df3.loc[~(df3['counts']<10)]
df4=df3.loc[~((df3['n']!=1) & (df3['m_sum']<5))]
df_final=df4.drop(['n','counts','m','m_sum'],axis=1)
df_final.reset_index()
df_final
    
```
